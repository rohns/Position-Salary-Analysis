```python
def cleanData():
    df.SalaryUSD =df.SalaryUSD.replace({'\$': '', ',': ''}, regex=True).astype(float)
    df.SalaryUSD = df.SalaryUSD.astype(int)
cleanData()
```

```python
df.head()
```

```python
%matplotlib inline
```

```python
db = df[df["PrimaryDatabase"].isin(["Microsoft SQL Server","Oracle","PostgreSQL"])]
```

```python
grouped = db.groupby(["Survey Year","PrimaryDatabase"]).mean()["SalaryUSD"]
grouped
```

```python
grouped2 = grouped.unstack(level="PrimaryDatabase")
grouped2.reset_index()
```

```python
grouped2.plot()
```

```python
top10 = df.groupby("Country")["SalaryUSD"].mean().sort_values().tail(10)
```

```python
top10
```

```python
top10.plot(x="Country", y= "SalaryUSD", kind="bar", title = "Top 10 mean Salaries for data positions (2017-2019)", ylabel = "Salary")
```

```python
db2 = df[-df["PrimaryDatabase"].isin(["Oracle","Microsoft SQL Server", "Other"])]
```

```python
db2.groupby("PrimaryDatabase")["PrimaryDatabase"].count().sort_values().tail(4).plot(kind="pie", ylabel = "", autopct='%1.1f%%', title= "Minor Database Popularity \n(Oracle, MS SQL Server Excluded)")
```
